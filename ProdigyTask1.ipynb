```python
# Import Important Libraries...

import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns  
```

```python
# Innsertion Of Data...

dataset= pd.read_excel(r"C:\Users\ALOK BHAGWAT\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_34.xlsx")
dataset
```

```python
# No of rows and columns in data 
dataset.shape
```

```python
# Information about dataset 
dataset.info()
```

```python
# Information about columns 
dataset.columns.tolist
```

```python
# Remove unnecessary columns ....
dataset= dataset.drop(columns=(['Country Code','Indicator Name','Indicator Code']))
dataset
```

```python
# Remove null values 
print(dataset.isnull().sum())
```

```python
# Filling The Null values...
dataset=dataset.fillna(0)
```

```python
# Give Name Of The Column 0
dataset.index.name = "ID NO"
dataset
```

```python
#Remove Null values data is clean ..... 
print(dataset.isnull().sum())
```

```python
# Convert data into tidy format 

dataset_tidy=dataset.melt(
    id_vars=['Country Name'],
    var_name='Year',
    value_name='Population'
)
dataset_tidy
```

```python
# Convert Year Data to numeric 
dataset_tidy['Year'] = pd.to_numeric(dataset_tidy['Year'])
dataset_tidy
```

```python
plt.figure(figsize=(8,5))
sns.histplot(dataset_tidy["Population"], bins=30, kde=True)
plt.title("Distribution of Population")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

This histogram visualizes the distribution of population values across different countries and years.
The KDE curve helps understand the density of values.
Most countries fall within lower population ranges, while a few countries have extremely high populations.

```python
# For populated countries...
aggregate_patterns = [
    "World",
    "income",
    "IDA",
    "IBRD",
    "dividend",
    "area",
    "countries",
    "classification",
    "HIPC",
    "fragile",
    "conflict",
    "OECD",
    "Asia",
    "Europe",
    "Africa",
    "Pacific",
    "America",
    "Arab",
    "Union",
    "Small states"
]
pattern = "|".join(aggregate_patterns)
```

```python
# Convert dataset into tidy for analysis 
dataset_tidy = dataset_tidy[
    ~dataset_tidy["Country Name"].str.contains(
        pattern, case=False, na=False
    )
]
```

```python
latest_year = dataset_tidy["Year"].max()
df_latest = dataset_tidy[dataset_tidy["Year"] == latest_year]
```

```python
# define 10 countries...
top10 = df_latest.sort_values(
    by="Population", ascending=False
).head(10)
```

```python
# For the 10 populated countries...
plt.figure(figsize=(10,6))
plt.bar(top10["Country Name"], top10["Population"])
plt.xticks(rotation=90, ha="right")
plt.xlabel("Country")
plt.ylabel("Population")
plt.title(f"Top 10 Most Populated Countries ({latest_year})")
plt.tight_layout()
plt.show()
```

This bar chart compares the maximum population of the top 10 most populated countries.


```python
# Average Popuulation Per Year Across The World...

avg_pop= dataset_tidy.groupby('Year')['Population'].mean()

plt.figure(figsize=(10,5))
avg_pop.plot(kind="bar")
plt.title('Avg Population Per Year Across The World...')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
```

This Indicates That The Population Increases Across Year By Year....

```python
#  For Particular India Population....

india_pop=dataset_tidy[dataset_tidy['Country Name']== "India"]

plt.figure(figsize=(10,5))
plt.bar(india_pop['Year'],india_pop['Population'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
```

This Shows That India Population Increases Year By Year Rapidaly...
